# Correlation Matrix
A correlation matrix is used to analyze the relationship between numerical variables in a dataset. It shows how strongly pairs of variables are related using correlation coefficients (ranging from -1 to 1).

## Interpreting the Correlation Matrix
Values close to +1 → Strong positive correlation 

Values close to -1 → Strong negative correlation 

Values close to 0 → Weak or no correlation 


```python
corr_mat=df.corr(numeric_only=True)
sns.heatmap(corr_mat,annot=True,cmap='Reds',fmt=".2f")
plt.title("Correlation Matrix")
```

We can observe from Correlation matrix that PetalWidth and PetalLength are very closely related as we are getting value equals to 0.96
